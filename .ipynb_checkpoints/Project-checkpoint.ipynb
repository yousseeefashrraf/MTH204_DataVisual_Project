## Mean & median & mode

# Mode

```python
def getMode(dataset): 
    allModes = []
    for iD in dataset:
        valueWithNumberOfOccurency = list(set(dataset[iD].sort_values()))
        index = 0
        for i in valueWithNumberOfOccurency:
            valueWithNumberOfOccurency[index] = (list(dataset[iD]).count(i),i)
            index += 1
        
        valueWithNumberOfOccurency.sort()
    
        maxOccur = valueWithNumberOfOccurency[len(valueWithNumberOfOccurency)-1]
        modes = (maxOccur[1],)
    
        for i in range(len(valueWithNumberOfOccurency)-2 , -1, -1):
            if valueWithNumberOfOccurency[i][0] == maxOccur[0]:
                modes += valueWithNumberOfOccurency[i][1],

        allModes.append({iD: modes})
    return allModes
getMode(ds)
```

# Mean

```python
def getMean(dataset):
    meanList = [0 for i in dataset]
    for i in range(0,len(dataset)):
        meanList += dataset.loc[i]
    return meanList/len(dataset)
print(getMean(ds))
```

# Median

```python
def getMed(dataset):
     if len(dataset)%2 == 0:
         return (dataset.loc[len(dataset)/2]+dataset.loc[len(dataset)/2+1])/2
     else :
         return dataset.loc[len(dataset)//2]
print(getMed(ds))
```

# Variance for each feature 

```python
def getVar(ds):
    variance = pd.DataFrame()
    variance['var(Pregnancies)'] = (ds['Pregnancies']-ds['Pregnancies'].mean())**2
    variance['var(Glucose)'] = (ds['Glucose']-ds['Glucose'].mean())**2
    variance['var(BloodPressure)'] = (ds['BloodPressure']-ds['BloodPressure'].mean())**2
    variance['var(SkinThickness)'] = (ds['SkinThickness']-ds['SkinThickness'].mean())**2
    variance['var(Insulin)'] = (ds['Insulin']-ds['Insulin'].mean())**2
    variance['var(BMI)'] = (ds['BMI']-ds['BMI'].mean())**2
    variance['var(DiabetesPedigreeFunction)'] = (ds['DiabetesPedigreeFunction']-ds['DiabetesPedigreeFunction'].mean())**2
    variance['var(Age)'] = (ds['Age']-ds['Age'].mean())**2
    variance['var(Outcome)'] = (ds['Outcome']-ds['Outcome'].mean())**2
    return variance.sum()/(len(variance)-1)
    
getVar(ds)
```

```python
def getSd(ds):
    variance = getVar(ds)
    SD = {

    }
    SD['SD(Pregnancies)'] = np.sqrt(variance['var(Pregnancies)'])
    SD['SD(Glucose)'] = np.sqrt(variance['var(Glucose)']) 
    SD['SD(BloodPressure)'] = np.sqrt(variance['var(BloodPressure)'])
    SD['SD(SkinThickness)'] = np.sqrt(variance['var(SkinThickness)'])
    SD['SD(Insulin)'] = np.sqrt(variance['var(Insulin)'])
    SD['SD(BMI)'] = np.sqrt(variance['var(BMI)'])
    SD['SD(DiabetesPedigreeFunction)'] = np.sqrt(variance['var(DiabetesPedigreeFunction)'])
    SD['SD(Age)'] = np.sqrt(variance['var(Age)'])
    SD['SD(Outcome)'] = np.sqrt(variance['var(Outcome)'])
    return SD
getSd(ds)
```
